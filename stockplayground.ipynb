```python
def filter_five_days(df):
    df = df.sort_values(by=['Date'], ascending=False)
    end_date= datetime.datetime.now() #- datetime.timedelta(1)
    start_date= end_date-datetime.timedelta(5)
    end_week_no = end_date.weekday()
    start_week_no = start_date.weekday()
    while end_week_no >= 5:
        end_date = end_date - datetime.timedelta(1)
        end_week_no = end_date.weekday()
    while start_week_no >= 5 :
        start_date = start_date - datetime.timedelta(1)   
        start_week_no = start_date.weekday()
    e = str(end_date)[:10]
    s = str(start_date)[:10]
    df1 = pd.DataFrame(columns=['Symbol', 'Five'])
    while True:
        try:
            for column in df:
                #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
                #print(df.index[0])
                percent_change = (df.loc[df.index[0], column] - df.loc[df.index[5], column])/df.loc[df.index[5], column]*100 #(df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
                df1.loc[len(df1)] = [column, round(percent_change,2)]

        except:
            start_date = start_date - datetime.timedelta(1)
            s = str(start_date)[:10]
        else:
            break
    return df1
```

```python
def export_to_xslx(df, file_name):
    try:
        #if file exists append the existing data
        with pd.ExcelWriter(file_name,mode='a', if_sheet_exists='overlay') as writer:  
            df.to_excel(writer, engine='openpyxl', sheet_name=file_name.split('.')[0], startrow = 10)
    except:
        #else create that file
        df.to_excel(file_name, sheet_name=file_name.split('.')[0], engine='openpyxl', startrow = 10)
    print('writing done') 
```

```python
def filter_year_date(df):
    df = df.sort_values(by=['Date'], ascending=False)
    end_date= datetime.datetime.now() #- datetime.timedelta(1)
    start_date= datetime.datetime(2022, 12, 30)
    end_week_no = end_date.weekday()
    start_week_no = start_date.weekday()
    while end_week_no >= 5:
        end_date = end_date - datetime.timedelta(1)
        end_week_no = end_date.weekday()
    while start_week_no >= 5:
        start_date = start_date - datetime.timedelta(1)   
        start_week_no = start_date.weekday()
    e = str(end_date)[:10]
    s = str(start_date)[:10]
    print(e)
    df1 = pd.DataFrame(columns=['Symbol', 'YTD'])
    while True:
        try:
            for column in df:
                #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
                percent_change = (df.loc[df.index[0], column] - df.loc[s, column])/df.loc[s, column]*100
                df1.loc[len(df1)] = [column, round(percent_change,2)]
        except:
            start_date = start_date - datetime.timedelta(1)
            s = str(start_date)[:10]
        else:
            break
    
    return df1
        
def filter_year(df):
    end_date= datetime.datetime(2022,12,30)
    start_date= datetime.datetime(2021,12,31)
    end_week_no = end_date.weekday()
    start_week_no = start_date.weekday()
    while end_week_no >= 5:
        end_date = end_date - datetime.timedelta(1)
        end_week_no = end_date.weekday()
    while start_week_no >= 5:
        start_date = start_date - datetime.timedelta(2)   
        start_week_no = start_date.weekday()
    e = str(end_date)[:10]
    s = str(start_date)[:10]
    df1 = pd.DataFrame(columns=['Symbol', '1Y CHANGE'])
    try:
        for column in df:
            #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
            percent_change = (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
            df1.loc[len(df1)] = [column, round(percent_change,2)]
    
    except:
        start_date = start_date - datetime.timedelta(1)
        s = str(start_date)[:10]
        for column in df:
            #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
            percent_change = (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
            df1.loc[len(df1)] = [column, percent_change]
    
    return df1

def filter_two_year(df):
    end_date = datetime.datetime(2021,12,31)
    start_date = datetime.datetime(2020,12,31)
    end_week_no = end_date.weekday()
    start_week_no = start_date.weekday()
    while end_week_no >= 5:
        end_date = end_date - datetime.timedelta(1)
        end_week_no = end_date.weekday()
    while start_week_no >= 5:
        start_date = start_date - datetime.timedelta(1)    
        start_week_no = start_date.weekday()
    e = str(end_date)[:10]
    s = str(start_date)[:10]
    df1 = pd.DataFrame(columns=['Symbol', '2Year CHANGE'])
    try:
        for column in df:
            #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
            percent_change = (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
            df1.loc[len(df1)] = [column, round(percent_change,2)]
    
    except:
        start_date = start_date - datetime.timedelta(1)
        s = str(start_date)[:10]
        for column in df:
            #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
            percent_change = (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
            df1.loc[len(df1)] = [column, percent_change]
    
    return df1

def filter_three_year(df):
    end_date = datetime.datetime(2020,12,31)
    start_date = datetime.datetime(2019,12,31)
    end_week_no = end_date.weekday()
    start_week_no = start_date.weekday()
    while end_week_no >= 5:
        end_date = end_date - datetime.timedelta(1)
        end_week_no = end_date.weekday()
    while start_week_no >= 5:
        start_date = start_date - datetime.timedelta(1)    
        start_week_no = start_date.weekday() 
    e = str(end_date)[:10]
    s = str(start_date)[:10]
    df1 = pd.DataFrame(columns=['Symbol', '3Year CHANGE'])
    try:
        for column in df:
            #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
            percent_change = (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
            df1.loc[len(df1)] = [column, round(percent_change,2)]
    
    except:
        start_date = start_date - datetime.timedelta(1)
        s = str(start_date)[:10]
        for column in df:
            #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
            percent_change = (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
            df1.loc[len(df1)] = [column, percent_change]
    
    return df1

def filter_four_year(df):
    end_date = datetime.datetime(2019,12,31)
    start_date = datetime.datetime(2018,12,31)
    end_week_no = end_date.weekday()
    start_week_no = start_date.weekday()
    while end_week_no >= 5:
        end_date = end_date - datetime.timedelta(1)
        end_week_no = end_date.weekday()
    while start_week_no >= 5:
        start_date = start_date - datetime.timedelta(1)    
        start_week_no = start_date.weekday()   
    e = str(end_date)[:10]
    s = str(start_date)[:10]
    df1 = pd.DataFrame(columns=['Symbol', '4Year CHANGE'])
    try:
        for column in df:
            #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
            percent_change = (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
            df1.loc[len(df1)] = [column, round(percent_change,2)]
    
    except:
        start_date = start_date - datetime.timedelta(1)
        s = str(start_date)[:10]
        for column in df:
            #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
            percent_change = (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
            df1.loc[len(df1)] = [column, percent_change]
    
    return df1

def filter_five_year(df):
    end_date = datetime.datetime(2018,12,31)
    start_date = datetime.datetime(2017,12,29)
    end_week_no = end_date.weekday()
    start_week_no = start_date.weekday()
    while end_week_no >= 5:
        end_date = end_date - datetime.timedelta(1)
        end_week_no = end_date.weekday()
    while start_week_no >= 5:
        start_date = start_date - datetime.timedelta(1)    
        start_week_no = start_date.weekday()     
    e = str(end_date)[:10]
    s = str(start_date)[:10]
    df1 = pd.DataFrame(columns=['Symbol', '5Year CHANGE'])
    try:
        for column in df:
            #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
            percent_change = (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
            df1.loc[len(df1)] = [column, round(percent_change,2)]
    
    except:
        start_date = start_date - datetime.timedelta(1)
        s = str(start_date)[:10]
        for column in df:
            #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
            percent_change = (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
            df1.loc[len(df1)] = [column, percent_change]
    
    return df1

def get_tickers():
    URL = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    table = soup.find('table', {'class': 'wikitable sortable'})
    tickers = []

    for row in table.findAll('tr')[1:]:
        #the format that wikipedia gives is slightly different than the one expected by yahoo finance, so change
        ticker = row.findAll('td')[0].text.replace("\n","").replace(".","-")
        tickers.append(ticker)
    return tickers
# Function to get ticker symbols for TSX60, returns list of strings
def get_tickers_tsx():
    URL = "https://en.wikipedia.org/wiki/S%26P/TSX_60"
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    table = soup.find('table', {'class': 'wikitable sortable'})
    tickers = []
    
    for row in table.findAll('tr')[1:]:
        # Put the tickers in the right format for yfinance api
        ticker = row.findAll('td')[0].text.replace("\n","").replace(".", "-")+".TO"
        tickers.append(ticker)
    return tickers
```

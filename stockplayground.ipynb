```python
print(yf.__version__)
def get_data(symbols, time_period):
    #print("Requesting information for " + name)
    #print(start_date, end_date)
    df = yf.download(tickers = symbols, period = time_period, threads=True)['Close']
    return df
df = get_data(symbols, '2y')

#filter_five_days("d")
```

```python
def filter_five_days(df):
    end_date= datetime.datetime.now() - datetime.timedelta(1)
    start_date= end_date-datetime.timedelta(5)
    end_week_no = end_date.weekday()
    start_week_no = start_date.weekday()
    if end_week_no > 6:
        end_date - datetime.timedelta(2)
    elif start_week_no > 6:
        start_date = datetime.timedelta(2)    
    e = str(end_date)[:10]
    s = str(start_date)[:10]
    df1 = pd.DataFrame(columns=['Symbol', 'Five'])
    try:
        for column in df:
            #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
            percent_change = (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
            df1.loc[len(df1)] = [column, percent_change]
    
    except:
        start_date = start_date - datetime.timedelta(1)
        s = str(start_date)[:10]
        for column in df:
            #print(column, 'Percent change', (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100)
            percent_change = (df.loc[e,column] - df.loc[s, column])/df.loc[s, column]*100
            df1.loc[len(df1)] = [column, percent_change]
    
    return df1
        
filtered = filter_five_days(df)
a = filtered.sort_values(by=['Five'],ascending=False)
a.head()
```
